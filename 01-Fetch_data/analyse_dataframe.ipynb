# Data ingestion using pandas

```python
df = pd.read_csv(DATA_SET_PATH, nrows=100000)
df
```

```python
df = df.rename(columns={
    "Duration in Minutes": "duration", 
    "Start Time": "start_datetime", 
    "End Time": "end_datetime",
    "Amount": "amount",
    "Kiosk ID": "kiosk_id",
    "App Zone ID": "app_zone_id",
    "App Zone Group" : "app_zone_group",
    "Payment Method": "payment_method",
    "Location Group": "location",
    "Last Updated": "modification_datetime"
})
df
```

```python
df.start_datetime = pd.to_datetime(df.start_datetime)
df.end_datetime = pd.to_datetime(df.end_datetime)
df.modification_datetime = pd.to_datetime(df.modification_datetime)
df.kiosk_id = df.kiosk_id.fillna(-1).astype(int)
df.app_zone_id = df.app_zone_id.fillna(-1).astype(int)
df.app_zone_group = df.app_zone_group.fillna("Unknown")
df["amount_per_hour"] = (df.amount / df.duration) * 60
df
```

# Data ingestion using pyspark

```python
import pyspark
from pyspark.sql import SparkSession, types
from pyspark.sql.functions import to_timestamp, col, round
```

```python
schema = types.StructType(
    [
        types.StructField('ID', types.IntegerType(), True),
        types.StructField('Source', types.StringType(), True),
        types.StructField('Duration in Minutes', types.FloatType(), True),
        types.StructField('Start Time', types.StringType(), True),
        types.StructField('End Time', types.StringType(), True),
        types.StructField('Amount', types.FloatType(), True),
        types.StructField('Kiosk ID', types.IntegerType(), True),
        types.StructField('App Zone ID', types.IntegerType(), True),
        types.StructField('App Zone Group', types.StringType(), True),
        types.StructField('Payment Method', types.StringType(), True),
        types.StructField('Location Group', types.StringType(), True),
        types.StructField('Last Updated', types.StringType(), True),
    ]
)
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('test') \
    .getOrCreate()

df = spark.read \
    .option("header", "true") \
    .schema(schema) \
    .csv(str(DATA_SET_PATH))

df.show(truncate=False)
```

```python
df = df.withColumnRenamed("ID", "id") \
        .withColumnRenamed("Source", "source") \
        .withColumnRenamed("Duration in Minutes", "duration_in_min") \
       .withColumnRenamed("Amount", "amount") \
       .withColumnRenamed("Kiosk ID", "kiosk_id") \
       .withColumnRenamed("App Zone ID", "app_zone_id") \
       .withColumnRenamed("App Zone Group", "app_zone_group") \
       .withColumnRenamed("Payment Method", "payment_method") \
       .withColumnRenamed("Location Group", "location")
```

```python
df = df.withColumn("start_datetime", to_timestamp("Start Time", "MM/dd/yyyy hh:mm:ss a")) \
.withColumn("end_datetime", to_timestamp("End Time", "MM/dd/yyyy hh:mm:ss a")) \
.withColumn("modification_datetime", to_timestamp("Last Updated", "MM/dd/yyyy hh:mm:ss a")) \
.withColumn("amount_per_hour", round((col("amount") / col("duration_in_min")) * 60, 2))
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df = df.repartition(12)
```

```python
df.select("id", "source", "modification_datetime", "start_datetime", "end_datetime", "duration_in_min", "amount", "amount_per_hour", "app_zone_id", "app_zone_group", "payment_method", "location") \
    .write.parquet('parking-transactions', mode="overwrite")
```
