```python
import pandas as pd
import pathlib
```

```python
DATA_SET = pathlib.Path.home() / "Traffic-fatalities-analyse" / "data" / "Parking_Transactions.csv"
```

```python
DATA_SET
```

```python
df = pd.read_csv(DATA_SET, nrows=5)
```

```python
df
```

```python
df = df.rename(columns={
    "Duration in Minutes": "duration", 
    "Start Time": "start_datetime", 
    "End Time": "end_datetime",
    "Amount": "amount",
    "Kiosk ID": "kiosk_id",
    "App Zone ID": "app_zone_id",
    "App Zone Group" : "app_zone_group",
    "Payment Method": "payment_method",
    "Location Group": "location",
    "Last Updated": "modification_datetime"
})
```

```python
df
```

```python
df.start_datetime = pd.to_datetime(df.start_datetime, format='%m/%d/%Y %I:%M:%S %p')
df.end_datetime = pd.to_datetime(df.end_datetime, format='%m/%d/%Y %I:%M:%S %p')
df.modification_datetime = pd.to_datetime(df.modification_datetime, format='%m/%d/%Y %I:%M:%S %p')
```
